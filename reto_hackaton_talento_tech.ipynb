```python
unidades_defectuosas=df[df['unidades_defectuosas']>0]
print(unidades_defectuosas.count())
```

```python
#eliminacion de columnas inesesarias
unidades_defectuosas=unidades_defectuosas.drop(columns=['tipo_fallo','observaciones'])
print(unidades_defectuosas.info())
```

```python
tabla_dinamica_unidades_def = pd.pivot_table(
    data=unidades_defectuosas,
    values='unidades_defectuosas',
    index='maquina_id',
    aggfunc='sum'
)

print(tabla_dinamica_unidades_def)
```

```python
tabla_dinamica_cantidad_prod=pd.pivot_table(data=unidades_defectuosas,values='cantidad_producida',index='maquina_id',aggfunc='sum')
print(tabla_dinamica_cantidad_prod)
```

```python
# calulado el porcentaje de unidades defectusas con respecto a la cantidad producida
tabla_dinamica_unidades_def['porcentaje_unidades_defectuosas']=(tabla_dinamica_unidades_def['unidades_defectuosas']/tabla_dinamica_cantidad_prod['cantidad_producida'])*100
print(tabla_dinamica_unidades_def)
```

```python
tabla_visual = tabla_dinamica_unidades_def.copy()
tabla_visual['cantidad_producida'] = tabla_dinamica_cantidad_prod['cantidad_producida']

print(tabla_visual)
```

```python
maquinas = tabla_dinamica_unidades_def.index.astype(str)  # Nombres o IDs de las máquinas
porcentajes = tabla_dinamica_unidades_def['porcentaje_unidades_defectuosas']

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(maquinas, porcentajes, color='salmon')  # Puedes cambiar el color a tu gusto

# Añadir detalles
plt.xlabel('Máquinas')
plt.ylabel('Porcentaje de Unidades Defectuosas (%)')
plt.title('Porcentaje de Defectos por Máquina')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar
plt.tight_layout()
plt.show()
'''
podemos llegar a la conclusion que las maquinas no son el problema de las unidades defetuosas

'''
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Asegúrate de que 'timestamp' es datetime
unidades_defectuosas['timestamp'] = pd.to_datetime(unidades_defectuosas['timestamp'])
unidades_defectuosas.set_index('timestamp', inplace=True)

# Agrupación mensual
mensual = unidades_defectuosas.resample('M').size()
mensual.index = mensual.index.to_period('M').to_timestamp()  # Opcional: poner al 1er día del mes

# (Opcional) Rellenar meses faltantes
rango_meses = pd.date_range(start=mensual.index.min(), end=mensual.index.max(), freq='MS')
mensual = mensual.reindex(rango_meses, fill_value=0)

# Filtramos hasta agosto
mensual_filtrado = mensual[ (mensual.index < '2023-09-01') ]

# Visualización
plt.figure(figsize=(12, 5))
plt.plot(mensual_filtrado.index, mensual_filtrado.values, marker='o', linestyle='-', color='teal')

# Agregamos los valores arriba de los puntos
for fecha, valor in zip(mensual_filtrado.index, mensual_filtrado.values):
    plt.text(fecha, valor + 5, str(valor), ha='center')

plt.title('Unidades defectuosas por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de defectos')
plt.grid(True)
plt.tight_layout()
plt.savefig('unidades_defectuosas_por_mes.png')
plt.show()
```

```python
# Paso 1: Convertir a DataFrame para trabajar con columnas
mensual_df = mensual.reset_index()
mensual_df.columns = ['fecha', 'cantidad_defectos']

# Paso 2: Crear tabla dinámica por número de mes
tabla_dinamica_mes = pd.pivot_table(
    data=mensual_df,
    values='cantidad_defectos',
    index=mensual_df['fecha'].dt.month,
    aggfunc='sum'
)

print(tabla_dinamica_mes)
```

```python
import calendar

# Paso 1: Convertir la serie mensual en DataFrame
mensual_df = mensual.reset_index()
mensual_df.columns = ['fecha', 'cantidad_defectos']

# Paso 2: Extraer número de mes y año
mensual_df['mes'] = mensual_df['fecha'].dt.month
mensual_df['año'] = mensual_df['fecha'].dt.year

# Filtrar para dejar solo meses anteriores a septiembre
mensual_df = mensual_df[ mensual_df['fecha'] < '2023-09-01' ]


# Paso 3: Calcular días del mes usando calendar
mensual_df['dias_en_mes'] = mensual_df.apply(
    lambda row: calendar.monthrange(row['año'], row['mes'])[1],
    axis=1
)

# Paso 4: Calcular defectos diarios promedio
mensual_df['defectos_por_dia'] = mensual_df['cantidad_defectos'] / mensual_df['dias_en_mes']

# Paso 5: Mostrar la tabla
print(mensual_df[['fecha', 'cantidad_defectos', 'dias_en_mes', 'defectos_por_dia']])

# Paso 6: Graficar
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(mensual_df['fecha'], mensual_df['defectos_por_dia'], marker='o', color='orange')
plt.title('Promedio diario de unidades defectuosas por mes')
plt.xlabel('Mes')
plt.ylabel('Defectos por día')
plt.grid(True)
plt.tight_layout()
plt.show()
```
